# FIFA WOMENS WORLD CUP - EDA 

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

**Loading & Reading data**

```python
df = pd.read_csv('womens-world-cup.csv')
```

#### Checking Duplicate & Missing Values

```python
df.isnull().sum()
```

*now, we fill the missing values*

```python
df['possesion'] = df['possesion'].fillna(0)
df['yellow_cards'] = df['yellow_cards'].fillna(0)
df['red_cards'] = df['red_cards'].fillna(0)
```

```python
df.duplicated().sum()
```

In the dataset there is not a single duplicate value.

### Data Visualization

```python
plt.figure(figsize = (20,8))
sns.heatmap(df.corr(), annot= True)
plt.show()
```

# Average of Squad Age

```python
plt.figure(figsize = (20,8))
sns.histplot(x = "age", data = df, color = 'navy')
plt.title("Histogram of Average Squad Age")
plt.xlabel('Age')
plt.xticks(rotation = 90)
sns.set_style('whitegrid')
```

```python
plt.figure(figsize = (20,8))
sns.boxplot(x = 'year',y = "age", data = df, color = 'red')
plt.title("Average Squad Age by Year")
plt.xlabel('Age')
plt.xticks(rotation = 90)
sns.set_style('dark')
plt.show()
```

# Total Goals scored by Squad (1991 to 2019)

```python
for j in df.year.unique():
    
    data0 = df[df['year'] == j]
    fig = px.bar(data0, x = 'squad', y = 'goals',
                color = 'goals',  color_continuous_scale = 'RdBu_r',
                title = f"Total goals scored by Squads in Each Year {j}")
    
    fig.update_layout(template = "plotly_dark")
    fig.show()
```

# Goals scored without Penalty by Squad (1991 to 2019)

```python
for j in df.year.unique():
    
    data0 = df[df['year'] == j]
    fig = px.bar(data0, x = 'squad', y = 'non_penalty_goals',
                color = 'goals',  color_continuous_scale = 'RdBu_r',
                title = f"Squad scored goals without Penalty {j}")
    
    fig.update_layout(template = "plotly_dark")
    fig.update_layout(yaxis_title_text= 'Non penalty goals')
    fig.show()
```

# Goals scored with Penalty by Squad (1991 to 2019)

```python
for j in df.year.unique():
    
    data0 = df[df['year'] == j]
    fig = px.bar(data0, x = 'squad', y = 'penalty_kicks_made',
                color = 'goals',  color_continuous_scale = 'RdBu_r',
                title = f"Squad scored goals with Penalty {j}")
    
    fig.update_layout(template = "plotly_dark")
    fig.update_layout(yaxis_title_text= 'Goals with penalty')
    fig.show()
```

# Total Possesion by Squad (2003 to 2019)

```python
for j in [2019, 2015, 2011, 2007, 2003]:
    
    data0 = df[df['year'] == j]
    fig = px.bar(data0, x = 'squad', y = 'possesion',
                color = 'possesion',  color_continuous_scale = 'RdBu_r',
                title = f"Total possesion by squad in {j}")
    
    fig.update_layout(template = 'plotly_dark')
    fig.update_layout(yaxis_title_text= 'Possesion Count')
    fig.show()
```

```python
match_per_year = df.groupby(['year'])['id'].count().reset_index().rename(columns={'id':'matches_played'})
match_per_year.style.background_gradient(cmap='PuBu')
```

```python
plt.figure(figsize = (20,7))
sns.barplot(x = match_per_year.year, y = match_per_year.matches_played, data = df)
plt.title("Number of matches played by each year")
plt.xlabel('year')
plt.ylabel('Count')
sns.set_style('whitegrid')
plt.show()
```

# Total Yellow Cards show to Squad (2015 - 2019)

```python
sns.set_context('notebook', font_scale = 1.3)
plt.figure(figsize=(22, 7))
ax = sns.barplot(x=df['squad'], 
                 y=df['yellow_cards'],
                 palette='viridis',
                 ci = None)
plt.ylabel('Yellow Cards');
plt.xticks(rotation = 90)

for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + 0.14, p.get_height() + 1), ha = 'center', va = 'bottom', color = 'Black')
```

# Total Red Cards show to Squad (2015 - 2019)

```python
sns.set_context('notebook', font_scale = 1.3)
plt.figure(figsize=(22, 7))
ax = sns.barplot(x=df['squad'], 
                 y=df['red_cards'],
                 palette='viridis',
                 ci = None)
plt.ylabel('Red Cards');
plt.xticks(rotation = 90)

for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + 0.14, p.get_height() + 1), ha = 'center', va = 'bottom', color = 'Black')
```
